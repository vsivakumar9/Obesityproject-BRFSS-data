```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
age_sex_2016 = pd.read_csv("ObesityAgeAndSex_2016.csv")
```

```python
age_sex_2016.count()
```

```python
age_sex_2016["BMI Category"] = age_sex_2016["BMI Category"].replace({"Normal Weight":"Others","Don't know/Refused/Missing":"Others",
                                                                     "Underweight":"Others","Overweight":"Others"})
age_sex_2016.head()
```

```python
age_sex_2016["BMI Category"].unique()
```

```python
state_obesity = age_sex_2016.groupby(["State","BMI Category"])["Date"].count()
```

```python
state_obesity = state_obesity.unstack(level=1)
```

```python
state_obesity["Total"] = state_obesity["Obese"] + state_obesity["Others"] 
state_obesity["Obesity %"] = state_obesity["Obese"] / state_obesity["Total"] * 100
```

```python
state_obesity_2016 = state_obesity
state_obesity_2016.head()
```

```python
state_obesity_2016.reset_index(inplace=True)
```

```python
state_obesity_2016.head()
```

```python
state_obesity_2016_df = state_obesity_2016[["State","Obesity %"]]
state_obesity_2016_df
```

```python
state_obesity_2016_df.to_csv("state_obesity_2016.csv",index=False)
```
