```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
import folium

pd.options.mode.chained_assignment = None

data2012 = pd.read_csv('FilteredObesity2012.csv')
data2013 = pd.read_csv('FilteredObesity2013.csv')
data2014 = pd.read_csv('FilteredObesity2014.csv')
data2015 = pd.read_csv('FilteredObesity2015.csv')
data2016 = pd.read_csv('FilteredObesity2016.csv')

data2012['Year'] = '2012'
data2013['Year'] = '2013'
data2014['Year'] = '2014'
data2015['Year'] = '2015'
data2016['Year'] = '2016'

data_all = pd.concat([data2012, data2013, data2014, data2015, data2016], ignore_index=True)

data2016.head()
```

```python
#Ethnicity Data: Seaborn Bar Plot of % Obese by All Respondents by Ethnicity

#1) Data Preparation
ethnic_bar_df = data_all.copy()

ethnic_bar_df['BMI Category'] = ethnic_bar_df['BMI Category'].replace({"Normal Weight": "Other", "Don't know/Refused/Missing": "Other","Underweight":"Other"})

group_bar = ethnic_bar_df.groupby(['Year', 'Race/Ethnicity', 'BMI Category'])['Date'].count()

final_bar_df = group_bar.unstack(level=2)

final_bar_df["Grand Total"] = final_bar_df["Obese"] + final_bar_df["Other"] + final_bar_df["Overweight"]

final_bar_df["Percentage of Obese (by Ethnicity)"] = (final_bar_df['Obese'] * 100) / final_bar_df['Grand Total']

final_bar_df.reset_index(inplace=True)

final_bar_df.head()
```

```python
## 2) Construct Bar Plot
import seaborn as sns

sns.set()

sns.set(font_scale=2)
sns.factorplot(x="Percentage of Obese (by Ethnicity)", y='Race/Ethnicity', hue='Year', data=final_bar_df, kind='bar', size=7, aspect=2)
eth_bar = plt.title("% of Obese Respondents Based on All Respondents of the Given Ethnicity")
plt.xlim(0, 100, 5)

plt.savefig('Visuals\ethnicity_obese_percentage.png')
plt.show()
```

```python
#Combined Data: Folium Chloropath Map of Frequency of Respondents by State, 2012

#1) Data Preparation
map_data = ('us-states.json')

data2012_map = data2012[['State', 'BMI Category', 'Race/Ethnicity']].copy()

data2012_grp = data2012.groupby('State')

data2012_totals = pd.DataFrame({"State": data2012_map['State'].unique(),
                              "Number of All Survey Respondents": data2012_grp['State'].count()})

data2012_totals = data2012_totals[['State', 'Number of All Survey Respondents']]

data2012_totals.head()
```

```python
#2) Construct Map
m2012 = folium.Map(location=[48, -102], zoom_start=3)

m2012.choropleth(
    geo_data=map_data,
    name='choropleth',
    data=data2012_totals,
    columns=['State', 'Number of All Survey Respondents'],
    key_on='feature.id',
    fill_color='YlGnBu',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Frequency of All Respondents, 2012'
)


folium.LayerControl().add_to(m2012)

m2012.save('Visuals\map_all_respondents_2012.htm')
m2012
```

```python
#Combined Data: Folium Chloropath Map of Frequency of Respondents by State, 2016
#1) Data Preparation
map_data = ('us-states.json')

data2016_map = data2016[['State', 'BMI Category', 'Race/Ethnicity']].copy()

data2016_grp = data2016.groupby('State')

data2016_totals = pd.DataFrame({"State": data2016_map['State'].unique(),
                              "Number of All Survey Respondents": data2016_grp['State'].count()})

data2016_totals = data2016_totals[['State', 'Number of All Survey Respondents']]

data2016_totals.head()

m2016 = folium.Map(location=[48, -102], zoom_start=3)

m2016.choropleth(
    geo_data=map_data,
    name='choropleth',
    data=data2016_totals,
    columns=['State', 'Number of All Survey Respondents'],
    key_on='feature.id',
    fill_color='YlGnBu',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Frequency of All Respondents, 2016'
)


folium.LayerControl().add_to(m2016)
m2016.save('Visuals\map_all_respondents_2016.htm')

m2016
```

```python
#Combined Data: % Of Obese Respondents by State in 2016
#1) Data Preparation
map_data = ('us-states.json')

obese_percent = pd.read_csv('state_obesity_2016.csv')

obese_map_2016 = folium.Map(location=[48, -102], zoom_start=3)

obese_map_2016.choropleth(
    geo_data=map_data,
    name='choropleth',
    data=obese_percent,
    columns=['State', 'Obesity %'],
    key_on='feature.id',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='% of Obese Respondents by State, 2016'
)


folium.LayerControl().add_to(obese_map_2016)
obese_map_2016.save('Visuals\map_obese_percentage_2016.htm')

obese_map_2016
```
