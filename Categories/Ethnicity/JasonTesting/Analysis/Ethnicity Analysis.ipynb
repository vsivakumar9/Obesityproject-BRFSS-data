```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt

pd.options.mode.chained_assignment = None

eth2016 = pd.read_csv('ObesityEthnicity2016.csv')

eth2016.head()
```

```python
eth2016['BMI Category'].value_counts()
```

```python
#delete all BMI Categories that aren't Obesity
dfobese = eth2016[eth2016['BMI Category'].str.contains("Obese") == True]
#change it to 1 for easy summing
dfobese['BMI Category'] = dfobese['BMI Category'].replace({"Obese":1})
                  
# dfobese['BMI Category'].value_counts()
dfobese.dtypes
```

```python
ethnicgroup = dfobese.groupby('Race/Ethnicity (General)')
```

```python
#pie chart: show overall ethnic counts by respondents
#Make sure BMI Category is Obesity Only

totalobese = ethnicgroup.sum()['BMI Category']

pielabels = totalobese.index

#chart
plt.figure(figsize=(7,7))
nationalobesity = plt.pie(totalobese, labels=pielabels, autopct="%1.2f%%", shadow=True, startangle=95, rotatelabels=True)

# plt.legend(loc="best", bbox_to_anchor=(1.1, 1.05))
plt.title("Overall Ethnic Obesity Rates in the US Counts by Respondents, 2016",  y=1.50)
plt.axis('equal')

plt.show()
```

```python
#pie chart: show ethnic counts by selected state only
#function takes in country code and returns pie chart for the state
def stateobesity(statecode):
    #select the state
    try:
        dfobese = eth2016[eth2016['BMI Category'].str.contains("Obese") == True]
        dfobese['BMI Category'] = dfobese['BMI Category'].replace({"Obese":1})
        
        dfstate = dfobese[dfobese['State'].str.contains(statecode) == True]

        ethnicgroup = dfstate.groupby('Race/Ethnicity (General)')

        totalobese = ethnicgroup.sum()['BMI Category']

        pielabels = totalobese.index

        #chart
        plt.figure(figsize=(7,7))
        stateobesity = plt.pie(totalobese, labels=pielabels, autopct="%1.2f%%", shadow=True, startangle=95, rotatelabels=True)

#         plt.legend(loc="best", bbox_to_anchor=(1.1, 1.05))
        plt.title(f"Overall Ethnic Obesity Rates in {statecode} State by Survey Respondents, 2016",  y=1.50)
        plt.axis('equal')

        return plt.show()
    
    except:
        print("Something went wrong!")
```
